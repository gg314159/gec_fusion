```python
#保存文件夹下图片的完整路径
import os
import pandas as pd
path = "/home/data/gecpet/AFFINE_REG_SKS_MASK_2D_2" #PET
#path = "/home/data/gec_1045mri_affine/AFFINE_REG_SKS_MASK_2D" #MRI
rootdir = path
i = 0
filelist = pd.DataFrame(columns=['filename'])
for parent, dirnames, filenames in os.walk(rootdir):
     for filename in filenames:
         i = i+1
         print("Parent folder:", parent)
         print("Filename:", filename)
         filelist.loc[i]=[parent+"/"+filename]


print("save csv ...")
filelist.to_csv("filelist_PET.csv", index=False)
```
