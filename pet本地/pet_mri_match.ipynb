```python
import pandas as pd
pet = pd.read_csv("2597pet_10_28_2022.csv")
mri = pd.read_csv("3505ADNI_CLEAN.csv")
```

```python
mri_pet=pd.merge(mri,pet,on="Subject")
```

```python
mri_pet.to_csv("mri_pet.csv")
```

```python
mri_pet_save = pd.DataFrame(columns=['Image Data ID_x', 'Subject', 'Group_x', 'Sex_x', 'Age_x', 'Visit_x',
       'Description_x', 'Acq Date_x', 'Image Data ID_y', 'Group_y', 'Sex_y',
       'Age_y', 'Visit_y', 'Modality', 'Description_y', 'Type', 'Acq Date_y',
       'Format', 'Downloaded'])
```

```python
for index, row in mri_pet.iterrows():
    age = abs(row['Age_x']-row['Age_y'])
    if age == 0:
        print(row['Image Data ID_y'])
        mri_pet_save.loc[index] = row
        mri_pet.drop(mri_pet.index[('Image Data ID_y'== row['Image Data ID_y'])],axis=0,inplace=True)
```

```python
mri_pet_save.to_csv("mri_pet_save.csv")
```
